```r
library(tidyverse)
library(forcats)
library(repr)
library(broom)
options(repr.matrix.max.rows = 6)
library(GGally)
```

```r
pm_data <- read.csv("data/pokemon.csv") 
pm_data
```

```r
set.seed(1234) 

## 1: standardize the data for clustering
pm_scaled <- pm_data %>% 
    select(Total:Speed) %>% 
    mutate(across(everything(), scale))
pm_scaled
    
```

```r
# lets use ggpairs

pm_eda <- pm_data%>% 
    select(Total:Speed) %>%
    ggpairs()
pm_eda
```

```r
pm_multi <- pm_scaled %>% 
    select(Total:Speed) %>% 
    mutate(across(everything(), scale))

pm_multi
```

```r
pm_multi_elbow_stats <- tibble(k = 1:10) %>% 
    rowwise() %>% 
    mutate(poke_clusts = list(kmeans(pm_multi, centers = k, nstart = 10)), 
           glanced = list(glance(poke_clusts)))

clustering_stats <- pm_multi_elbow_stats %>% 
    select(-poke_clusts) %>% 
    unnest(glanced) 

clustering_stats
```

```r
elbow_plot <- clustering_stats %>% 
    ggplot(aes(x = k, y = tot.withinss)) +
    geom_point() + 
    geom_line() +
    labs(x = "K", y = "Total within-cluster sum of squares", title = "Elbow Plot") +
    scale_x_continuous(breaks = 1:10) + 
    theme(text = element_text(size = 20)) 

elbow_plot
```

```r
new_pm_scaled <- pm_scaled %>% 
    select(Attack, Defense) %>% 
    as_tibble()

new_pm_scaled 
```

```r
clusters <- kmeans(new_pm_scaled, centers = 3) 
clusters
```

```r
clustered_pm <- augment(clusters, new_pm_scaled) 
clustered_pm
```

```r
cluster_plot <- clustered_pm %>% 
    ggplot(aes(x = Attack, y = Defense, color = .cluster)) +
    geom_point() + 
    labs(x = "Attack", y = "Defense", color = "Cluster") +
    theme(text = element_text(size = 12))
cluster_plot
```

```r
glance(clusters)
```

```r
poke_clust_ks <- tibble(k = 1:10) %>%
    rowwise() %>% 
    mutate(clusters = list(kmeans(new_pm_scaled, k)), 
          glanced = list(glance(clusters))) 

cluster_stats <- poke_clust_ks %>%
    unnest(glanced) 

```

```r
pm_plot <- cluster_stats %>% 
    ggplot(aes(x = k, y = tot.withinss)) + 
    geom_point() + 
    geom_line() +
    labs(x = "K", y = "Total Within-cluster sum of squares", title = "Elbow plot") + 
    theme(text = element_text(size = 12)) 

pm_plot
```
