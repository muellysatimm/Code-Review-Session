```r
library(tidyverse)
library(repr)
library(tidymodels)
library(cowplot)
options(repr.matrix.max.rows = 6)
source("tests.R")
source('cleanup.R')
```

```r
marathon <- read.csv("data/marathon.csv") 
```

```r
marathon
```

```r
set.seed(1) 
marathon_split <- initial_split(marathon, prop = 0.75, strata = time_hrs) # strata is what we're trying to predict 
marathon_train <- training(marathon_split) 
marathon_test <- testing(marathon_split)

```

```r
# visualize the plot to see which variable has a correlation 

marathon_plot <- marathon_train %>% 
    ggplot(aes(x = max, y = time_hrs)) + 
    geom_point() + 
    labs(x = "Body Mass Index",
            y = "Time in Hours")+ 
    theme(text = element_text(size=12)) 

marathon_plot
```

```r
# lets create the spec model 
lm_spec <- linear_reg() %>% 
    set_engine("lm") %>% 
    set_mode("regression")

lm_spec
```

```r
lm_recipe <- recipe(time_hrs ~ max, data = marathon_train) 
lm_recipe
```

```r
# lets fit the model and recipe into the workflow()

lm_fit <- workflow() %>% 
    add_model(lm_spec) %>% 
    add_recipe(lm_recipe) %>% 
    fit(data = marathon_train) 

lm_fit
```

predicted time vs. distance = B0 + B1 

= 4.8661 - 0.0216


```r
# lets add the line of best fit 
lm_predictions <- ggplot(marathon_train, aes(x = max, y = time_hrs)) +
    geom_point() + 
    labs(x = "Maximum Distance per week", 
         y = "Race time in hours", 
        title = "Time vs Distance") +
    geom_smooth(method = "lm", se = FALSE) + 
    theme(text = element_text(size=12))

lm_predictions
```

```r
# lets calculate the RMSPE on the test set

lm_test_result <- lm_fit %>% 
                predict(marathon_test) %>% 
                bind_cols(marathon_test) %>% 
                metrics(truth = time_hrs, estimate = .pred)

lm_test_result
    
```

```r
# lets pull the rmspe 

lm_rmspe <- lm_test_result%>% 
    filter(.metric == "rmse") %>% 
    select(.estimate) %>%
    pull()

lm_rmspe
```

```r
# lets vizualize it 

lm_predictions_test <- marathon_test %>% 
                    ggplot(aes(x = max, y = time_hrs)) +
                    geom_point(alpha = 0.25, size = 2) + 
                    geom_smooth(data = marathon_train, aes(x = max, y = time_hrs), method = "lm", se = FALSE) +
                    labs(x = "Maximum distance per Week", y = "Race Time (hrs)") + 
                    theme(text = element_text(size = 20)) + 
                    ggtitle("Relationship between Race Time and Max Distance") 

lm_predictions_test
```

```r
credit <- read.csv("data/Credit.csv")
credit
```

```r
new_credit <- credit %>% 
        select(Income, Rating, Balance) %>% 
        as_tibble()

new_credit
```

```r
# split data 

credit_split <- initial_split(new_credit, prop = 0.6, strata = Balance) 
credit_train <- training(credit_split) 
credit_test <- testing(credit_split) 
```

```r
credit_spec <- linear_reg() %>% 
            set_engine("lm") %>% 
            set_mode("regression") 

credit_recipe <- recipe(Balance ~ Income + Rating, data = credit_train) 

credit_spec
credit_recipe
```

```r
# fit the model 

credit_fit <- workflow() %>% 
    add_model(credit_spec) %>% 
    add_recipe(credit_recipe) %>% 
    fit(data = credit_train) 

credit_fit
```

```r
# lets calculate and predict the rmspe

credit_result <- credit_fit %>% 
                predict(credit_test) %>% 
                bind_cols(credit_test) %>% 
                metrics(truth = Balance, estimate = .pred) 
credit_result
```

```r
credit_rmspe <- credit_result %>% 
        filter(.metric == "rmse") %>% 
        select(.estimate) %>% 
        pull()

credit_rmspe
```

```r
# plot the data 

credit_plot <- credit_test %>% 
        ggplot(aes(x = Income, y = Balance)) + 
        geom_point() +
        labs(x = "Income", y = "Credit Card Balance") + 
        geom_smooth(method = "lm", se = FALSE) + 
        theme(text = element_text(size=12)) 

credit_plot
```

```r
new_credit_recipe <- recipe(Balance ~ Income, data = credit_train) 

new_credit_fit <- workflow() %>% 
    add_model(credit_spec) %>%
    add_recipe(new_credit_recipe) %>% 
    fit(data = credit_train) 

new_credit_fit
```

Balance Prediction = B0 + B1 * (income) 

= 239.45 + 6.094 * (income)
